```python
from matplotlib import pyplot as plt
from py_2048_game import Game
from py_2048_game import solvers
```

```python
solver = solvers.get_solver('py_2048_game.solvers.FirstActionSolver')()
game = Game(keep_history=False)

iterations = 1000

scores = []
moves = []
```

```python
for i in range(iterations):
    for _ in solver.solve_game(game):
        pass
    scores.append(game.score)
    moves.append(game.move_count)
    game.reset()
```

```python
figure, axis = plt.subplots(2, 1)

axis[0].hist(scores, bins=100)
axis[0].set_title('Score histogram')
axis[1].hist(moves, bins=100)
axis[1].set_title('Move histogram')
```

```python
plt.scatter(scores, moves, alpha=.1)
```
